```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
from sklearn import linear_model,datasets
import numpy as np
import math
import random
```

```python
#logistics regression to predict class missing values

iris = datasets.load_iris()
iris_data=iris.data
iris_target=iris.target
df_label=pd.DataFrame(iris_data)
df_label[4] = pd.Series(iris_target, index=df_label.index,dtype='float')

ix = [(row, 4) for row in range(df_label.shape[0])]
for row,col in random.sample(ix, int(round(.1*len(ix)))):
    df_label.iat[row, col] = np.nan

new_df_label=df_label.dropna()
count=0
for iter1 in df_label:
    if count==0:
       k1_label=df_label[pd.isnull(df_label[iter1])]
    else:
       k1_label=k1_label.append(df_label[pd.isnull(df_label[iter1])])    
    count=count+1

h=new_df_label.as_matrix()
logreg = linear_model.LogisticRegression(C=1e5)

X_all=new_df_label.iloc[:,0:4].copy() 
y=new_df_label[4].as_matrix()
y=np.reshape(y,(len(y),1))
logreg.fit(X_all,y)


k_modi_1=k1_label.copy()
for enum,i in enumerate(k1_label.index.values):
    if(np.isnan(k1_label[4][i])):
        k_modi_1[4][i]=logreg.predict(k1_label.iloc[enum,0:4].reshape(1,4))

        
        
for i in range(y.shape[0]):
    if(y[i][0]==0):
        plt.scatter(X_all.iloc[i,0],X_all.iloc[i,2], c='red')
    elif(y[i][0]==1):
        plt.scatter(X_all.iloc[i,0],X_all.iloc[i,2], c='blue')
    elif(y[i][0]==2):
        plt.scatter(X_all.iloc[i,0],X_all.iloc[i,2], c='green')

for i in range(k_modi_1.shape[0]):
    if(k_modi_1.iloc[i,4]==0):
        plt.scatter(k_modi_1.iloc[i,0],k_modi_1.iloc[i,2], s=50,c='red',marker ="D")
    elif(k_modi_1.iloc[i,4]==1):
        plt.scatter(k_modi_1.iloc[i,0],k_modi_1.iloc[i,2], s=50,c='blue',marker ="D")
    elif(k_modi_1.iloc[i,4]==2):
        plt.scatter(k_modi_1.iloc[i,0],k_modi_1.iloc[i,2], s=50,c='green',marker ="D")
        
plt.show()        
```

```python

iris = datasets.load_iris()
iris_data=iris.data
iris_target=iris.target
df_label=pd.DataFrame(iris_data)
df_label[4] = pd.Series(iris_target, index=df_label.index,dtype='float')

ix = [(row, 3) for row in range(df_label.shape[0])]
for row,col in random.sample(ix, int(round(.1*len(ix)))):
    df_label.iat[row, col] = np.nan

new_df_label=df_label.dropna()
count=0
for iter1 in df_label:
    if count==0:
       k1_label=df_label[pd.isnull(df_label[iter1])]
    else:
       k1_label=k1_label.append(df_label[pd.isnull(df_label[iter1])])    
    count=count+1

h=new_df_label.as_matrix()

corr_matrix = new_df_label.corr()

max1=0
for iter1 in corr_matrix:
    if(iter1!=3):
       if corr_matrix[iter1][3]>max1:
           max1=corr_matrix[iter1][3]
           key_max=iter1

X=new_df_label[key_max].as_matrix()
y=new_df_label[3].as_matrix()


X=np.reshape(X,(len(X),1))
y=np.reshape(y,(len(y),1))

reg1=linear_model.LinearRegression()
reg1.fit(X, y)

k_modi=k1_label.copy()
for i in k1_label.index.values:
    if(np.isnan(k1_label[3][i])):
        k_modi[3][i]=reg1.predict(k1_label[key_max][i])
x_line=np.linspace(0,8,20)
y_line=reg1.coef_[0]*x_line+reg1.intercept_[0]
plt.plot(x_line, y_line, color='green', linewidth=2)        
plt.scatter(X,y, c='blue')        
plt.scatter(k_modi_1[key_max],k_modi_1[3], c='red')       
plt.show()        
   
```
