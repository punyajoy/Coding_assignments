```python
#skewedness of data
import math
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#symmetric data
values = range(1, 209)
probs = [1.0 / 208] * 208
 
for idx, prob in enumerate(probs):
    if idx > 70 and idx < 90:
        probs[idx] = probs[idx] * (1 + math.log(idx + 1))
    if idx > 90 and idx < 110:
        probs[idx] = probs[idx] * (1 + math.log((110 - idx) + 1))
 
probs = [p / sum(probs) for p in probs]
sample =  np.random.choice(values, 1000, p=probs)
binwidth = 2

mode=stats.mode(sample)
mean=np.mean(sample)
median=np.median(sample)
plt.hist(sample, bins=np.arange(min(sample), max(sample) + binwidth, binwidth))
plt.xlim([0, max(sample)])
plt.vlines(mean, 0, 50, colors='k',label='mean')
plt.vlines(median, 0, 50, colors='g',label='median')
plt.vlines(mode[0], 0, 50, colors='m',label='mode')
plt.legend()
plt.show()
```
